```python
import os
import face_recognition
import pandas as pd
```

```python
folder_path = "/Users/gennacontino/Desktop/ai-news-investigations-project-1/images"
```

```python
landmarks_data = []
```

```python
for filename in os.listdir(folder_path):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        image_path = os.path.join(folder_path, filename)

        image = face_recognition.load_image_file(image_path)

        face_landmarks_list = face_recognition.face_landmarks(image)

        if face_landmarks_list:
            for i, face_landmarks in enumerate(face_landmarks_list):
                landmarks_data.append({'filename': filename, 'face_id': i + 1, **face_landmarks})

            print(f"Found {len(face_landmarks_list)} faces in {filename}")
        else:
            print(f"No faces found in {filename}")
```

```python
df = pd.DataFrame(landmarks_data)
```

```python
csv_filename = "face_landmarks.csv"
df.to_csv(csv_filename, index=False)
```

```python
def find_min_max_coordinate_list(coordinate_list, x_or_y='x', min_or_max='min'):
    coordinates = coordinate_list
    if x_or_y == 'x':
      sort_key = 0
    else:
      sort_key = 1
    if min_or_max == 'min':
      return min(coordinates, key=lambda x: x[sort_key])[sort_key]
    else:
      return max(coordinates, key=lambda x: x[sort_key])[sort_key]
```

```python
df['nose_bridge_start'] = df['nose_bridge'].apply(lambda cl: find_min_max_coordinate_list(cl, 'y', 'max'))
```

```python
df['nose_bridge_end'] = df['nose_bridge'].apply(lambda cl: find_min_max_coordinate_list(cl, 'y', 'min'))
```

```python
df['left_eye_interior'] = df['left_eye'].apply(lambda cl: find_min_max_coordinate_list(cl, 'x', 'max'))
```

```python
df['right_eye_interior'] = df['right_eye'].apply(lambda cl: find_min_max_coordinate_list(cl, 'x', 'min'))
```

```python
df['nose_tip_start'] = df['nose_tip'].apply(lambda cl: find_min_max_coordinate_list(cl, 'x', 'min'))
```

```python
df['nose_tip_end'] = df['nose_tip'].apply(lambda cl: find_min_max_coordinate_list(cl, 'x', 'max'))
```

```python
df['bottom_lip_start'] = df['bottom_lip'].apply(lambda cl: find_min_max_coordinate_list(cl, 'x', 'min'))
```

```python
df['bottom_lip_end'] = df['bottom_lip'].apply(lambda cl: find_min_max_coordinate_list(cl, 'x', 'max'))
```
